```python
df = yf.download(tickers, start='2017-01-01')
df
```

### 3. Read 'Adj Close' price only from tickers list

```python
df1 = df['Adj Close']
df1
```

### 4. Plot 'Adj Close' price in the tickers list

```python
df1.plot()
```

### 5. Plot daily return rate of the list in tickers

```python
# Fetching the Adj Close prices and get its history and ETF
# Shifting to one day and fill the first with 0

df_dpc=df1/df1.shift(1)-1
df_dpc.iloc[0]=0

df_dpc
```

```python
df_dpc.plot()
```

```python
# Calculation of Cumulative return
df2=df_dpc.cumsum()

# Plot cumulative returns with pyplot
df2.plot()
```

```python
There are few popular ways to get stock price from yahoo.com.
```

### 1. Using pdr.DataReader

```python
#Fetch stack data from the starting data indicated to current day
#You can indicate the day of end as, end='      '.

#from pandas_datareader import data as pdr 
import yfinance as yf
# 

tickers = ['msft', 'aapl', 'twtr', 'amd', 'tsm', 'goog', 'amzn', 'fb', 'nvda']
#df = pdr.DataReader(tickers, data_source='yahoo', start='2017-01-01')
df = yf.download(tickers, data_source='yahoo', start='2017-01-01')

df
```

### 2. Using yf.download
It is simpler than the above, without data_source='yahoo'.

```python
df = yf.download(tickers, start='2017-01-01')
df
```

### 3. Using pdr.get_data_yahoo

```python
#pip install pandas_datareader
```

```python
import pandas_datareader as pdr
df = pdr.get_data_yahoo(tickers, start='2017-01-01')
df
```

### 4. Read 'Adj Close' price only from tickers list

```python
df1 = df['Adj Close']
df1
```

### 5. Plot 'Adj Close' price in the tickers list

```python
df1.plot()
```

### 6. Plot daily return rate of the list in tickers

```python
# Fetching the Adj Close prices and get its history and ETF
# Shifting to one day and fill the first with 0

df_dpc=df1/df1.shift(1)-1
df_dpc.iloc[0]=0

df_dpc
```

```python
df_dpc.plot()
```

```python
# Calculation of Cumulative return
df2=df_dpc.cumsum()

# Plot cumulative returns with pyplot
df2.plot()
```
